```python
movie_id_list = []
movie_title = []
movie_release_date=[]
movie_description=[]
movie_rating=[]
```

```python
url_test="https://www.themoviedb.org/movie/top-rated"
get(url_test)
```

```python
uClint = uReq(url_test)
html_page = uClint.read()
uClint.close()
```

```python
page_soup = soup(html_page,"html.parser")
```

```python
movies_container = page_soup.findAll("div",{"class":"item poster card"})
```

```python
for x in range(0, len(movies_container)):
    movie_details = movies_container[x].find("div",{"class":"image_content"})
    movie_id = re.findall('\d+',movie_details.a["id"])
    movie_id_list.append(movie_id[0])
    movie_title.append(movie_details.a["title"])
    movie_release_date.append(movies_container[x].find("div",{"class":"info"}).find("div",{"class":"wrapper"}).find("div",{"class":"flex"}).find("span").text)
    movie_description.append(movies_container[x].find("div",{"class":"info"}).find("p",{"class":"overview"}).text.replace("…",""))
    movie_rating.append(movies_container[x].find("div",{"class":"info"}).find("div",{"class":"consensus tight"}).find("div",{"class":"outer_ring"}).div["data-percent"])
```

```python
data_dictonary = {'Movie ID':movie_id_list,
                  'Movie title':movie_title,
                  'Movie release date':movie_release_date,
                  'movie Description':movie_description,
                  'movie rating':movie_rating
                  }
```

```python
data_frame_webscrapper = pd.DataFrame(data= data_dictonary)
```

```python
movie_id_list = [19404,278,238,372058,324857,424,240,129,497,637,452522,680,18491,539,155,550,26451,13,346,40096]
```

```python
movie_budget = []
movie_revenue = []
movie_runtime = []
movie_genres = []
movie_popularity =[]
movie_genre = ""
movie_profit =[]
movie_cost_per_min=[]
```

```python
#Looping the Movie ID's and getting the data and appending them to the corresponding lists created in the above line
for x in range(0, len(movie_id_list)):
   movie = tmdb.Movies(movie_id_list[x])
   info = movie.info()
   budget_int =int( movie.budget / 100000)
   movie_budget.append(budget_int)
   revenue_int = int(movie.revenue / 100000)
   movie_revenue.append(revenue_int)
   profit = revenue_int - budget_int
   movie_profit.append(profit)
   movie_runtime.append(movie.runtime)
   cost_per_min = float(budget_int/movie.runtime)
   movie_cost_per_min.append(cost_per_min)

   temp_list = movie.genres
   for i in range(0,len(temp_list)):
       temp_dict = temp_list[i]
       movie_genre = movie_genre + temp_dict['name'] + ','
   movie_genres.append(movie_genre)
   movie_genre = ""

   movie_popularity.append(movie.popularity)
```

```python
#Creating a data dictionary to store the values

data_dictonary = {'Movie Budget (in Hundred thousands)':movie_budget,
                 'Movie Revenue (in Hundred thousands)':movie_revenue,
                 'Movie Runtime (in minutes)':movie_runtime,
                 'Movie Genres':movie_genres,
                 'Movie popularity':movie_popularity,
                 'Movie Profit (in Hundred thousands)':movie_profit,
                 'Movie CostPerMin (in Hundred thousands per minute)':movie_cost_per_min}
```

```python
data_frame_api = pd.DataFrame(data= data_dictonary)
```

```python
#Creating CSV Dataset

from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from requests import get
import pandas as pd
import re
```

```python
directors = []
producers = []
screenplay = []
cast = []
```

```python
list_of_urls = ['https://en.wikipedia.org/wiki/Dilwale_Dulhania_Le_Jayenge','https://en.wikipedia.org/wiki/The_Shawshank_Redemption',
                'https://en.wikipedia.org/wiki/The_Godfather','https://en.wikipedia.org/wiki/Your_Name','https://en.wikipedia.org/wiki/Spider-Man:_Into_the_Spider-Verse',
                'https://en.wikipedia.org/wiki/Schindler%27s_List','https://en.wikipedia.org/wiki/The_Godfather_Part_II',
                'https://en.wikipedia.org/wiki/Spirited_Away','https://en.wikipedia.org/wiki/The_Green_Mile_(film)','https://en.wikipedia.org/wiki/Life_Is_Beautiful',
                'https://en.wikipedia.org/wiki/Twin_Peaks','https://en.wikipedia.org/wiki/Pulp_Fiction','https://en.wikipedia.org/wiki/The_End_of_Evangelion',
                'https://en.wikipedia.org/wiki/Psycho_(1960_film)','https://en.wikipedia.org/wiki/The_Dark_Knight_(film)',
                'https://en.wikipedia.org/wiki/Fight_Club','https://en.wikipedia.org/wiki/Investigation_of_a_Citizen_Above_Suspicion',
                'https://en.wikipedia.org/wiki/Seven_Samurai','https://en.wikipedia.org/wiki/Forrest_Gump',
                'https://en.wikipedia.org/wiki/Goodfellas'
                ]
```

```python
#Getting the data for the required attributes by scraping WikiPedia source.
for x in range(0, len(list_of_urls)):
    uClint = uReq(list_of_urls[x])
    html_page = uClint.read()
    uClint.close()
    soup_page = soup(html_page, "html.parser")

    container = soup_page.findAll("table",{"class":"infobox vevent"})

    contain = container[0].find("tbody").findAll("tr")

    try:
        d_names = contain[2].findAll("td")
        cleaned = d_names[x].text.replace("\n",";").replace("[a]","").replace("[1]","")
        test = cleaned[0]
    except:
        d_names = contain[4].findAll("td")
    else:
        d_names = contain[2].findAll("td")
    directors_sub = []
    for x in range(0,len(d_names)):
        if (re.match(r'[[a-z]+]', d_names[x].text)):
            pass
        else:
            cleaned = d_names[x].text.replace("\n",";").replace("[a]","").replace("[1]","")
        if(cleaned[0]==';'):
            directors_sub.append(cleaned[1:])
        else:
            directors_sub.append(cleaned)

        temp_d = list(directors_sub)
    directors.append(temp_d[0])



    producers_sub = []

    p_names = contain[3].findAll("td")
    for x in range(0,len(p_names)):
        if (re.match(r'[[a-z]+]', p_names[x].text)):
            pass
        else:
            cleaned = p_names[x].text.replace("\n",";").replace("[a]","").replace("[1]","").replace("?","")
            if(cleaned[0]==';'):
                producers_sub.append(cleaned[1:])
            else:
                producers_sub.append(cleaned)
        temp_p = list(producers_sub)
    #print(temp_p)
    producers.append(temp_p[0])

    screenplay_sub = []

    s_names = contain[4].findAll("td")
    for x in range(0,len(s_names)):
        if (re.match(r'[[a-z]+]', s_names[x].text.replace("\n"," "))):
            pass
        else:
            cleaned = s_names[x].text.replace("\n",";").replace("[a]","").replace("[1]","")
            if(cleaned[0]==';'):
                screenplay_sub.append(cleaned[1:])
            else:
                screenplay_sub.append(cleaned)
        temp_s = list(screenplay_sub)
    #print(temp_s)
    screenplay.append(temp_s[0])


    cast_sub = []


    if (contain[7].find("th").text=='Starring'):
        c_names = contain[7].findAll("td")
    else:
        if (contain[8].find("th").text=='Starring'):
            c_names = contain[8].findAll("td")
        else:
            if (contain[10].find("th").text=='Starring'):
                c_names = contain[10].findAll("td")
            else:
                c_names = contain[6].findAll("td")
    for x in range(0,len(c_names)):
        if (re.match(r'[[a-z]+]', c_names[x].text)):
            pass
        else:
            cleaned = c_names[x].text.replace("\n",";").replace("[a]","").replace("[1]","")
        if(cleaned[0]==';'):
            cast_sub.append(cleaned[1:])
        else:
            cast_sub.append(cleaned)
        temp_c = list(cast_sub)
    #print(len(temp_c))
    cast.append(temp_c[0])
```

```python
data_dictonary = {'Director':directors,
                  'Producer':producers,
                  'Cast':cast,
                  'Screen':screenplay}
```

```python
data_frame_csv = pd.DataFrame(data= data_dictonary)
```
